```python
import pandas as pd

# Load raw GTFS data
stops = pd.read_csv('../data/raw/stops.txt')
stop_times = pd.read_csv('../data/raw/stop_times.txt')

# CLEANING STEP 1: Filter for Parent Stations (location_type = 1)
# These are your 'Nodes' for the cascade analysis
hubs = stops[stops['location_type'] == 1][['stop_id', 'stop_name', 'stop_lat', 'stop_lon']]

# CLEANING STEP 2: Handle 'stop_times' 
# The 'arrival_time' sometimes goes above 24:00:00 (e.g., 25:30) for late-night trains.
# We need to fix this for proper temporal analysis.
def fix_mta_time(time_str):
    h, m, s = map(int, time_str.split(':'))
    if h >= 24:
        h = h - 24
    return f"{h:02d}:{m:02d}:{s:02d}"

stop_times['arrival_time_clean'] = stop_times['arrival_time'].apply(fix_mta_time)

# CLEANING STEP 3: Save the 'Processed' version
hubs.to_csv('../data/processed/cleaned_hubs.csv', index=False)
stop_times.head(1000).to_csv('../data/processed/clean_stop_times_sample.csv', index=False)

print("Midpoint Deliverable 1: Data Cleaned and Processed!")
```
